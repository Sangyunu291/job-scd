```python
""" Circular_List는 무조건 List_Pointer를 통해서만 접근하기"""
class Circular_List :
    def __init__(self) :
        self.__c_list = []

    def append(self, x) :
        self.__c_list.append(x)

    def get_at(self, idx) :
        idx = idx % len(self.__c_list)
        return self.__c_list[idx]

    def sort(self) :
        pass

    def length(self) :
        return len(self.__c_list)

class List_Pointer :
    def __init__(self, p_list : Circular_List, idx) :
        self.__idx = idx
        self.__p_list = p_list

    def get_val(self) :
        temp = self.__p_list.get_at(self.__idx)
        self.__idx += 1
        return temp
```

# [체이닝 해시 테이블]


```python
""" 체이닝 해시 테이블 """
class ChainingHashTable:
    def __init__(self, size=10):
        self.size = size
        # 각 버킷을 빈 리스트로 초기화 (체이닝을 위한 공간)
        self.table = [[] for _ in range(self.size)]

    def _hash_function(self, key):
        # 파이썬 내장 hash 함수를 사용하여 인덱스 계산
        return hash(key) % self.size

    def set(self, key, value):
        """데이터 삽입/수정 (Set)"""
        index = self._hash_function(key)
        bucket = self.table[index]

        # 이미 키가 존재하는지 확인하여 업데이트
        for i, (k, v) in enumerate(bucket):
            if k == key:
                bucket[i] = (key, value)
                return

        # 키가 없으면 새로 추가
        bucket.append((key, value))

    def get(self, key):
        """데이터 조회 (Get)"""
        index = self._hash_function(key)
        bucket = self.table[index]

        for k, v in bucket:
            if k == key:
                return v
        return None  # 키가 없을 경우

    def has_key(self, key):
        """키 존재 여부 확인 (Contains)"""
        index = self._hash_function(key)
        bucket = self.table[index]

        for k, v in bucket:
            if k == key:
                return True
        return False

    def print_table(self):
            """해시 테이블 출력"""
            for c_list in self.table:
                for x in c_list :
                  k, v = x
                  print(f"key:{k}, value:{v}")
```
